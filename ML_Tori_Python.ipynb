```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os

# Any results you write to the current directory are saved as output.
# FILE_DIR = '../input/hawaiiml-data'
# for f in os.listdir(FILE_DIR):
#     print('{0:<30}{1:0.2f}MB'.format(f, 1e-6*os.path.getsize(f'{FILE_DIR}/{f}')))

FILE_DIR = '../Sprint09alt_Machine_Learning_Hawaii_Kaggle_Competition'
# for f in os.listdir(FILE_DIR):
#     print('{0:<30}{1:0.2f}MB'.format(f, 1e-6*os.path.getsize(f'{FILE_DIR}/{f}')))
    
df_train = pd.read_csv(f'{FILE_DIR}/train.csv', encoding='ISO-8859-1') #write training data to dataframe
df_test = pd.read_csv(f'{FILE_DIR}/test.csv', encoding='ISO-8859-1') # Read the test data
```

```python
df_train = df_train.assign(quantity_log1p = np.log1p(df_train.quantity))
df_train = df_train.loc[(df_train['unit_price'] > 0) & (df_train['unit_price'] < 10000) ]
df_train = df_train.assign(unit_price_log1p = np.log1p(df_train.unit_price))
```

```python
# created mean for unit price - going to do min and max as well.
#train %>% group_by(stock_id) %>% summarise(min(unit_price))
```

```python
# mutated the data to add mean  of unit price to the dataset
df_train["mean_price"] = df_train.groupby('stock_id')['unit_price'].transform(np.mean)           
```

```python
# mutated the data to add max of unit price to the dataset. 
df_train["max_price"] = df_train.groupby('stock_id')['unit_price'].transform(np.max)       
```

```python
# mutated the data to add min of unit price to the dataset. 
df_train["min_price"] = df_train.groupby('stock_id')['unit_price'].transform(np.min)        
```

```python
# mutated the data to add median of unit price to the dataset. 
df_train["median_price"] = df_train.groupby('stock_id')['unit_price'].transform(np.median)       
```

```python
df_train
```

```python
# subtracted unit price from median price to get the difference. 
#train = train %>% mutate(median_diff = unit_price - median_price)
df_train["median_diff"] = df_train["unit_price"] - df_train["median_price"]
```

```python
# subtracted unit price from mean price to get the difference. 
#train = train %>% mutate(mean_diff = unit_price - mean_price)
df_train["mean_diff"] = df_train["unit_price"] - df_train["mean_price"]
```

```python
# subtracted unit price from min price to get the difference. 
#train = train %>% mutate(min_diff = unit_price - min_price)
df_train["min_diff"] = df_train["unit_price"] - df_train["min_price"]
```

```python
# subtracted max price from unit price to get the difference. 
#train = train %>% mutate(max_diff = max_price - unit_price)
df_train["max_diff"] = df_train["max_price"] - df_train["unit_price"]
```
